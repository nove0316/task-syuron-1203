### HTMLの作成

```python
# 必要なライブラリのインポート
import pandas as pd
from jinja2 import Environment, FileSystemLoader
import os

# ディレクトリの設定
base_dir = os.getcwd()

template_dir = os.path.join(base_dir, 'templates')
data_dir = os.path.join(base_dir, 'data')
output_dir = os.path.join(base_dir, 'X')

# 出力ディレクトリが存在しない場合は作成
os.makedirs(output_dir, exist_ok=True)

# Jinja2環境の設定
env = Environment(loader=FileSystemLoader(template_dir))
template = env.get_template('template.html')

# lancersResultIdをキーとした辞書を作成
evidence_dict = splited_data_time.set_index('lancersResultId')['Q4-1'].to_dict()

# 各行に対してHTMLファイルを生成
for _, row in final_summarized_df_time_morning.iterrows():
    id = row['ID']
    proposition = row['proposition']
    # lancersResultIdsを整数のリストに変換
    lancers_ids = [int(x.strip()) for x in row['lancersResultIds'].split(',')]
    
    # 証拠リストを取得
    evidence_list = [evidence_dict.get(lid, f"証拠ID {lid} は存在しません") for lid in lancers_ids]
    
    # テンプレートにデータを渡してレンダリング
    html_content = template.render(id=id, proposition=proposition, evidence_list=evidence_list)
    
    # HTMLファイルのパスを設定
    output_file = os.path.join(output_dir, f'task_{id}.html')
    
    # HTMLファイルを書き出し
    with open(output_file, 'w', encoding='utf-8') as f:
        f.write(html_content)
    
    print(f'Generated {output_file}')
```
